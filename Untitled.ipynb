```python
p_range = list(range(1,8))  # taking values from 1 to 4
q_range = list(range(1,4))  # taking values from 1 to 8
aic_values = []
pq_values = []
d = 1
for p in p_range:
    for q in q_range:
        try:
            model = sm.tsa.arima.ARIMA(df, order=(p, 1, q),enforce_stationarity=False)
            results = model.fit()
            aic_values.append(results.aic)
            pq_values.append((p, q))
        except Exception as e:
            logging.error(traceback.format_exc())


best_pq = pq_values[aic_values.index(min(aic_values))]  # (p,q) corresponding to lowest AIC score
print("(p,q) corresponding to lowest AIC score: ", best_pq)
```

```python
def getP_Q(data, difference):
    p_range = list(range(1,8))  # taking values from 1 to 4
    q_range = list(range(1,8))  # taking values from 1 to 8
    aic_values = []
    pq_values = []
    d = difference
    for p in p_range:
        for q in q_range:
            try:
                model = sm.tsa.arima.ARIMA(data, order=(p, d, q),enforce_stationarity=False)
                results = model.fit()
                aic_values.append(results.aic)
                pq_values.append((p, q))
            except Exception as e:
                logging.error(traceback.format_exc())


    best_pq = pq_values[aic_values.index(min(aic_values))]  # (p,q) corresponding to lowest AIC score
    print("(p,q) corresponding to lowest AIC score: ", best_pq)
    return best_pq[0], best_pq[1]

a,b = getP_Q(df,2)
a,b
```

```python
arima_model = sm.tsa.arima.ARIMA(df, order=(best_pq[0], 1, best_pq[1]),enforce_stationarity=False).fit()
predictions = arima_model.predict(start=0, end=len(df)-1)

mae = mean_absolute_error(list(df['value']), list(predictions))
print("Mean Absolute Error:", mae)
```

## Smoothing and solving

```python
window =5
rolling_mean = df.rolling(window=window).mean()
rolling_mean
```

```python
rolling_mean
plt.plot(rolling_mean, 'g', label='Rolling mean window ')
plt.plot(df[window:])
plt.title("Smooth curved")
```

```python
diff = rolling_mean.diff().dropna()
plt.plot(diff)
X = diff
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
df_smooth = rolling_mean.dropna()
```

```python
p_range = list(range(1,20))
q_range = list(range(1,8))
aic_values = []
pq_values = []
# fitted_model_dict = {}
# count = 0
d = 1
for p in p_range:
    for q in q_range:
        try:
            model = sm.tsa.arima.ARIMA(df_smooth, order=(p, d, q),enforce_stationarity=False)
            results = model.fit()
            aic_values.append(results.aic)
            pq_values.append((p, q))
        except Exception as e:
            logging.error(traceback.format_exc())


best_pq = pq_values[aic_values.index(min(aic_values))]  # (p,q) corresponding to lowest AIC score
print("(p,q) corresponding to lowest AIC score: ", best_pq)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

arima_model = sm.tsa.arima.ARIMA(df_smooth, order=(best_pq[0], 0, best_pq[1]),enforce_stationarity=False).fit()
predictions = arima_model.predict(start=0, end=len(df_smooth)-1)

mae = mean_absolute_error(list(df_smooth['value']), list(predictions))
print("Mean Absolute Error:", mae)
```
